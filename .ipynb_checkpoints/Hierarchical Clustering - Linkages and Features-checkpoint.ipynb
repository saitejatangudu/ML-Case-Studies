## Hierarchical Clustering

Hierarchical Clustering: Linkages and Features. This task requires you to implement Hierarchical clustering, and perform clustering on a given dataset(s). The list of subtasks is given below. You are ex- pected to implement the required, using classes and methods. We expect to see routines like hc.linkages(X, linkage type) (takes the data and provides linkage matrix), hc.dendogram(Z) (takes the linkage matrix and plots a dendogram).

```python
class HC:
    def __init__(self,d_metric = 'euclidean'):
        self.clusters = []
        self.d_metric = d_metric
    def create_linkage(self, x, linkage_type='SL'):
        self.linkage_type = linkage_type
        linkage = []
        clusters = [[i] for i in range(len(x))]  # Initialize clusters as individual data points
        while len(clusters) > 1:
            fix_i = fix_j = 0
            min_dist = np.inf
            for i in range(len(clusters)):
                for j in range(i + 1, len(clusters)):
                    distance = self.cal_pair_distance(x[clusters[i][0]], x[clusters[j][0]])
                    if distance < min_dist:
                        fix_i = i
                        fix_j = j
                        min_dist = distance
        
        # Append the linkage information
            linkage.append([clusters[fix_i][0], clusters[fix_j][0], min_dist, len(clusters[fix_i]) + len(clusters[fix_j])])
        
        # Merge clusters
            clusters[fix_i].extend(clusters[fix_j])
            clusters.pop(fix_j)
        return linkage
    def cal_pair_distance(self,x,y):
        point_distances = np.zeros((len(x),len(y)))
        for i in range(len(x)):
            for j in range(len(y)):
                distance = np.linalg.norm(x[i]-y[j])
                point_distances[i][j] = distance 
        if(self.linkage_type =='SL'):
            return np.min(point_distances)
        if(self.linkage_type =='CL'):
            return np.max(point_distances)
        if(self.linkage_type =='GAL'):
            return np.mean(point_distances)
```

# Using SciPy

```python
from scipy.cluster.hierarchy import linkage
from scipy.cluster.hierarchy import dendrogram
import matplotlib.pyplot as plt
```

```python
linkage_data = linkage(scaled_data)
```

```python
linkage_data
```

```python
plt.figure()
plt.xlabel("Distance")
plt.ylabel('Clusters')
plt.title("Single Linkage")
dn =dendrogram(linkage_data,orientation='left')
```

```python
linkage_data = linkage(scaled_data,method='complete')
```

```python
linkage_data
```

```python
plt.figure()
plt.xlabel("Distance")
plt.ylabel('Clusters')
plt.title("Complete Linkage")
dn =dendrogram(linkage_data,orientation='left')
```

```python
linkage_data = linkage(scaled_data,method='average')
```

```python
linkage_data
```

```python
plt.figure()
plt.xlabel("Distance")
plt.ylabel('Clusters')
plt.title("Average Linkage")

dn =dendrogram(linkage_data,orientation='left')
```

```python
linkage_data = linkage(scaled_data,method='centroid')
```

```python
linkage_data
```

```python
plt.figure()
plt.xlabel("Distance")
plt.ylabel('Clusters')
plt.title("Centroid Linkage")

dn =dendrogram(linkage_data,orientation='left')
```

* Cluster Structures :
Complete linkage have formed clusters which have completely different structure then the ones formed by others. Since, It produced compact clusters then the rest
* Similar Clusters : 
Centroid linkage and Average linkage almost produced similar clusters in terms of the heights of the clusters
* Sensitivity to Outliers :
Single Linkage have produced elongated clusters this can be attributed to the outliers in the data whose effect got nullified with Complete Linkage
